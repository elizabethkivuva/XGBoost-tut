```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
X, y =data.drop('median_house_value', axis=1), data['median_house_value']
```

```python
# split the data we are going to use

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
#instantiate our XGBoost model
#fit a reg. model because we are predicting price of a house and our target column continous and dont expect values to be the same but keep on changing

xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 100)
```

```python
#run the fit

xg_reg.fit(X_train,y_train)
preds = xg_reg.predict(X_test)
```

```python
# run score of our model

xg_reg.score(X_test, y_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
# fit aa random forest model to compare the acuracy of regression model above
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
preds_rf = rf.predict(X_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, preds_rf))
print('Model Score:', str(rf.score(X_test, y_test)))
print("RMSE: %f" % (rmse_rf))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
preds_lr = lr.predict(X_test)
rmse_lr = np.sqrt(mean_squared_error(y_test, preds_lr))
print('Model Score:', str(lr.score(X_test, y_test)))
print("RMSE: %f" % (rmse_lr))
```

it has better accuracy of 81.3% compared to XGBoost model with 74.1%. However XGboost is beeter compared to linear regression with 61%
